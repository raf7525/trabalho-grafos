```python
from pathlib import Path
import pandas as pd


base_path = Path('./usa_airports_original_dataset.csv')
df = pd.read_csv(base_path)

df['Fly_date'] = pd.to_datetime(df['Fly_date'])
df_filtered = df[df['Fly_date'].dt.year.between(2005, 2009)]

df_filtered.head()

df_filtered.to_csv('./usa_airport_dataset.csv', index=False)
```
